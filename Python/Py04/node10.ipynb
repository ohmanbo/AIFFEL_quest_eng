# 키오스크 만들기

```python
from datetime import datetime

def ask_valid_number(question: str, valid_numbers: list[int]) -> int:
    while True:
        try:
            user_input = input(f"{question} ")
            selected_num = int(user_input)
            
            # 원하는 입력값인지 확인
            if selected_num in valid_numbers:
                return selected_num  
            else:
                print(f"{valid_numbers} 중에서 다시 입력해주세요.")
        
        except ValueError:
            # 숫자가 아닌 문자를 입력했을 때 예외 처리
            print("숫자만 입력가능 합니다.")

class Kiosk:
    # 시스템에 키오스크 기계가 여러개있다고 가정하고 주문번호가 겹치지않게 클래스변수로
    # Kiosk의 인스턴스가 매장 내 키오스크들중 하나.
    order_number = 0 
    
    def __init__(self):
        self.menu = ["Americano", "Latte", "Mocha", "YuzaTea", "GreenTea", "ChocoLatte"]
        self.price = [2000, 3000, 4000, 3000, 3000, 3500]
        self.order_menu = []    # 현재 주문 리스트
        self.order_price = []   # 가격 리스트

    # 메뉴 출력 메서드
    def menu_print(self):
        total_width = 51
        side_width = (total_width - 1) // 2 

        print('─' * total_width)
        print(f"{'C O F F E E':^{side_width}}|{'T E A':^{side_width}}")
        print('─' * total_width)
        left_menu = f" 1. {self.menu[0]:<14}{self.price[0]:>6} "
        right_menu = f" 4. {self.menu[3]:<14}{self.price[3]:>6} "
        print(f"{left_menu}|{right_menu}")
        left_menu = f" 2. {self.menu[1]:<14}{self.price[1]:>6} "
        right_menu = f" 5. {self.menu[4]:<14}{self.price[4]:>6} "
        print(f"{left_menu}|{right_menu}")
        left_menu = f" 3. {self.menu[2]:<14}{self.price[2]:>6} "
        right_menu = f" 6. {self.menu[5]:<14}{self.price[5]:>6} "
        print(f"{left_menu}|{right_menu}")
        print('─' * total_width)

    # 주문 메서드
    def menu_select(self):
        print()  # 줄 바꿈
        want_order = True
        while want_order:
            n = ask_valid_number("음료 번호를 입력하세요: ", [1,2,3,4,5,6])
            
            t = ask_valid_number("🍵HOT 음료는 [1]을, 🧊ICE 음료는 [2]를 입력하세요 : ", [1,2])
            if t == 1:
                temp = "🍵HOT"
            elif t == 2:
                temp = "🧊ICE"
            
            print('선택한 음료', temp, self.menu[n-1], ' : ', self.price[n-1], '원')  # 메뉴

            self.order_price.append(self.price[n-1])
            self.order_menu.append(temp + ' ' + self.menu[n-1])
                        
            print()  # 줄 바꿈
            self.print_selected_menu()

            # 추가주문 확인하기
            want = int(input("추가 주문을 원하시면 1을 입력, 결제하려면 아무거나 입력 : "))  # 음료 번호 입력
            if want == 1:
                want_order = True
                self.menu_print()
            else:
                want_order = False
    
    # 선택된 모든 음료를 출력해주는 함수
    def print_selected_menu(self):
        print()
        print(f"선택하신 음료 {len(self.order_menu)}개, 총 {sum(self.order_price)}원 입니다")
        print('─' * 51)
        

    # 지불
    def pay(self):
        p = ask_valid_number("지불방법을 입력하세요. 1: 현금, 2: 카드, 3: 카카오페이", [1,2,3])
        if p == 1:      print("직원을 호출했습니다. 직원에게 현금을 지불하세요")
        elif p == 2:    print("IC칩 방향으로 카드를 꽂아주세요")
        elif p == 3:    print("카카오페이를 선택하셨습니다.")
        else:           print("잘못 입력하셨습니다. 다시 입력 해주세요")

        Kiosk.order_number += 1;

    # 주문서 출력 
    def table(self):
        total_width = 31
        side_width = (total_width - 1) // 2 

        now = datetime.now()
        date_string = now.strftime("%Y-%m-%d %H:%M:%S")

        print("\n") # 2줄
        print('─' * total_width)
        print(f"{f'주문번호 {Kiosk.order_number:03d}':^{total_width}}") #가운데정렬
        print('─' * total_width)
        length = len(self.order_menu)
        for i in range(length):
            left_text = f"{self.order_menu[i]:<20}"
            right_text = f"{f'{self.order_price[i]} 원':>8}"
            print(f"{left_text} {right_text}")
        print('-' * total_width)
        left_text = f"{'총주문금액':<15}"
        right_text = f"{f'{sum(self.order_price)} 원':>11}"
        print(f"{left_text} {right_text}")
        print("\n") # 2줄

        left_text = f"{'주 문 담 당 자:':<{side_width-5}}"
        right_text = f"{'키오스크 1':>{side_width-2}}"
        print(f"{left_text} {right_text}")

        left_text = f"{'고    객    수:':<{side_width-3}}"
        right_text = f"{'1':>{side_width}}"
        print(f"{left_text} {right_text}")

        left_text = f"{'주문 고유 번호:':<{side_width-6}}"
        right_text = f"{Kiosk.order_number:>{side_width+1}}"
        print(f"{left_text} {right_text}")
        
        left_text = f"{'주 문 일 시:':<{side_width-8}}"
        right_text = f"{date_string:>{side_width+4}}"
        print(f"{left_text} {right_text}")
        print('─' * total_width)
```

```python
a = Kiosk()  # 객체 생성 
a.menu_print()  # 메뉴 출력
a.menu_select()  # 주문
a.pay()  # 지불
a.table()  # 주문표 출력
```
