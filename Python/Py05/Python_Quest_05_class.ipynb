```python
import random

class Account:
  account_list = [] # 여러 계좌를 저장
  account_count = 0
  def __init__(self, name: str, balance: int):
    # Q1. 은행이름과 계좌번호 자동설정, 초기잔액과 예금주는 입력받음
    self.bank = "SC은행"
    self.account_number = self._account_number_generator()
    self.name = name
    self.balance = balance

    # Q7. 이자
    self.deposit_count_for_interest = 0

    # Q10. 입출금 history
    self.deposit_log = []
    self.withdraw_log = []

  def _account_number_generator(self) -> str:
    # 각 섹션별로 랜덤한 숫자 조합 생성
    part1 = random.randint(1, 999)    # 3자리
    part2 = random.randint(1, 99)     # 2자리
    part3 = random.randint(1, 999999) # 6자리

    account_number = f"{part1:03}-{part2:02}-{part3:06}"

    # 중복된 계좌번호를 생성하지 않게 만들면 더 좋지만 시간 관계상 skip
    Account.account_count += 1
    Account.account_list.append(self)

    return account_number

  def get_account_num(): return Account.account_count

  # Q6
  def display_info(self): return print(f"은행이름:{self.bank} {self.name}님의 {self.account_number} 계좌에 총 잔액 {self.balance}원 입니다\n")

  # Q4
  def deposit(self, amount: int):
    if(amount <= 0):
      print("입금액은 0보다 커야합니다.")
      return
    else:
      self.balance += amount
      message = f"[deposit] {amount}원을 계좌에 입금 합니다"
      self.deposit_log.append(message)
      print(message)
      self._process_interest()                  # <- 이자 처리
      self.display_info()
      return

  # Q5
  def withdraw(self, amount: int):
    if(amount <= 0):
      print("출금액은 0보다 커야합니다.")
      return
    elif(self.balance < amount): # <- 계좌의 잔고 이상으로 출금 불가능
      print("잔액이 부족합니다.")
      return
    else:
      self.balance -= amount
      message = f"[withdraw] {amount}원을 출금 합니다"
      self.withdraw_log.append(message)
      print(message)
      self.display_info()
      return

  # Q7
  def _process_interest(self):
    self.deposit_count_for_interest += 1
    if(self.deposit_count_for_interest >= 5):
      amount = int(self.balance * 0.01)
      self.balance = self.balance + amount     # 1% 이자 지급
      self.deposit_count_for_interest = 0
      message = f"[deposit_interest] 이자로 {amount}원이 계좌에 지급 되었습니다."
      self.deposit_log.append(message)
      print(message)

  # Q9 100만원 이상 고객정보만 출력하기
  @staticmethod
  def display_VIP():
    print("VIP 고객 정보 (100만원이상 예금 보유자)")
    for acc in Account.account_list:
      if acc.balance >= 1000000:
        acc.display_info()

  # Q10
  def deposit_history(self):
    print("입금내역을 출력합니다.")
    for log in self.deposit_log:
      print(log)

  def withdraw_history(self):
    print("출금내역을 출력합니다.")
    for log in self.withdraw_log:
      print(log)







```
