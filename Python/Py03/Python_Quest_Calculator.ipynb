```python
# 연산 함수 정의
def add(a,b): return a+b
def sub(a,b): return a-b
def mul(a,b): return a*b
def div(a,b): return a/b
def pow(a,b): ## 제곱함수 직접 구현
  original_value = a
  for i in range(b-1):
    if b==0: return 1
    original_value *= a
  return original_value

# 정수 및 연산자 입력받기 함수
def ask_number():
  return input("정수를 입력하세요: ")
def ask_operator():
  print("1: + 더하기", "2: - 빼기", "3: * 곱하기", "4:  / 나누기", "5: ** 제곱\n", sep='\n')
  return input("보기 중에서 연산자를 고르세요 [정수 입력 후 엔터]: ")

# 유효성 검사 함수
def validate_operator(operator):
  # 입력받은 앖이 숫자가 1, 2, 3, 4, 5 중 하나인지 확인후,
  # 유효한값이면 그대로 리턴 아니면 None을 리턴 -> 재입력받기
  valid_options = [1, 2, 3, 4, 5]
  try:
    # 문자열 입력을 숫자로 변환
    num = int(operator)
    if num in valid_options:
      return num
  except ValueError:
    print(f"입력하신 {operator}은(는) 선택지에 없습니다. (1~5 사이 입력, 종료: Ctrl+C)")
    return None
  except:
    print("알수없는 에러 발생")
    return None

  print(f"입력하신 {operator}은(는) 선택지에 없습니다. (1~5 사이 입력, 종료: Ctrl+C)")
  return None

def validate_number(number, operator=None):
    # 입력값이 정수인지 확인하고,
    # operator가 나누기 연산일 경우 0이 아닌지도 함께 검증합니다.
    try:
      num = int(number)
      # operator_code가 4(나누기)이고, 입력값이 0인 경우만 ZeroDivisionError 발생
      if operator == 4 and num == 0:
        raise ZeroDivisionError

    except ValueError:
      print(f"오류: '{number}'은(는) 숫자가 아닙니다.")
      return None
    except ZeroDivisionError:
      print(f"오류: '{number}'으로 나눌수 없습니다")
      return None
    else:
      return num

def calculator():
  # 작동 원리
  # 1. 첫째 항, 연산자, 둘째 항을 순서대로 입력받고 실행
  # 2. 이후엔 연산자, 둘째 항 을 반복적으로 입력받아 Windows의 계산기처럼 연속적으로 계산
  # 3. first_value는 첫 계산이후 계산결과 누적
  # 4. operator와 second_value는 None값으로 초기화 하면서 연속해서 받아들이기
  # 4. 도중에 예외처리

  first_value = None
  operator = None
  second_value = None

  # 매번 입력시 예외처리 하고나서 실제 변수에 값을 넣기위한 임시 변수
  input_value = None

  # 첫째 항 입력받기 - 정수가 정상적으로 입력될때까지 반복
  while first_value is None:
    input_value = ask_number()
    result = validate_number(input_value)
    if result is not None:
      first_value = result
    else:
      continue

  # 메인 루프문 - (연산자 > 두번째 항) 반복해서 입력받기
  while True:
    ## 연산자 입력받기
    if operator is None:
      result = validate_operator(ask_operator())
      if result is not None:
        operator = result
      continue

    ## 두번째 항 입력받기
    if second_value is None:
      # 나누기일 때 0 체크를 위해 현재 operator 전달
      result = validate_number(ask_number(), operator)
      if result is not None:
        second_value = result
      else:
        continue # 잘못 입력했으면 다시 숫자 받기

    if operator is not None and second_value is not None:
      match operator:
        case 1: first_value = add(first_value, second_value)
        case 2: first_value = sub(first_value, second_value)
        case 3: first_value = mul(first_value, second_value)
        case 4: first_value = div(first_value, second_value)
        case 5: first_value = pow(first_value, second_value)

    print(f"\n[ 계산 결과: {first_value} ]\n")
    operator = None
    second_value = None
    continue # 연산자와 두번째항 초기화 후 루프의 처음으로 돌아감


calculator()
```
